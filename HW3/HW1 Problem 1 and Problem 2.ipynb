```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D

def make3d_plot_multivariate_normal(mu_x,variance_x,mu_y,variance_y,covariance,angle):
    x = np.linspace(-10,10,500)
    y = np.linspace(-10,10,500)
    X, Y = np.meshgrid(x,y)
    pos = np.empty(X.shape + (2,))
    pos[:, :, 0] = X; pos[:, :, 1] = Y
    rv = multivariate_normal([mu_x, mu_y], [[variance_x, covariance], [covariance, variance_y]])
    print(type(rv))

    # Make a 3D plot
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.view_init(azim=angle)
    ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=0)
    ax.set_xlabel('X axis')
    ax.set_ylabel('Y axis')
    ax.set_zlabel('Z axis')
    plt.show()
```

```python
#distribution 1 3d head on
make3d_plot_multivariate_normal(0,16,0,9,0,0)
```

```python
#distribution 1 3d at 30 degree angle
make3d_plot_multivariate_normal(0,16,0,9,0,30)
```

```python
#Part a:
from sklearn import linear_model
from sklearn.metrics import mean_squared_error
reg = linear_model.LinearRegression()
reg = reg.fit(X,y)
print('Coefficients of features in entire dataset')
print(reg.coef_)

#getting MSE
y_pred = reg.predict(X)
print('MSE entire dataset: ' + str(mean_squared_error(y,y_pred)))
```
